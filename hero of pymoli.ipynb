```python
import pandas as pd
import numpy as np
file_to_load = "Resources/purchase_data.csv"
purchase_data_pd = pd.read_csv(file_to_load)
purchase_data_pd.head()
```

```python
total_players= purchase_data_pd["SN"].nunique
total_players()
```

```python
item_count = purchase_data_pd["Item ID"].nunique()
avg_price = purchase_data_pd["Price"].mean()
purchase_count= purchase_data_pd["Purchase ID"].nunique()
total_revenue= purchase_data_pd["Price"].sum()
purchaseing_analysis = pd.DataFrame({"Number of Unique Items": [item_count], "Average Price": [avg_price], "Number of Purchases": [purchase_count], "Total Revenue": [total_revenue]})
purchaseing_analysis
```

```python
gender_table= purchase_data_pd.groupby(["Gender"])
gender_count= gender_table["SN"].nunique()
percent_gender= gender_count/total_players()
purchase_gender = pd.DataFrame({"Total Count": gender_count, "Percetage of Players": percent_gender})
purchase_gender
```

```python
purchasegender= purchase_data_pd.groupby(["Gender"])
item_gender= purchasegender["Item ID"].nunique()
avg_gender= purchasegender["Price"].mean()
purchcount_gender= purchasegender["Purchase ID"].nunique()
totalrev_gender= purchasegender["Price"].sum()
purchaseing_gender= pd.DataFrame({"Purchase Count": purchcount_gender, "Average Purchase Price": avg_gender, "Total Purchase Value": totalrev_gender, "Avg Total Purchase per Person": purchcount_gender})
purchaseing_gender
```

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 1000]
bins_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
purchase_data_pd["Age Group"]= pd.cut(purchase_data_pd["Age"], bins, labels=bins_labels)
age_purchase_data= purchase_data_pd.groupby("Age Group")
age_count=age_purchase_data["SN"].nunique()
perage_count = age_count/total_players()
age_demo = pd.DataFrame({"Total Count": age_count, "Percetage of Players": perage_count})
age_demo
```

```python
avg_age= age_purchase_data["Price"].mean()
purchcount_age= age_purchase_data["Purchase ID"].nunique()
totalrev_age= age_purchase_data["Price"].sum()
purchase_age= pd.DataFrame({"Purchase Count": purchcount_age, "Average Purchase Price": avg_price, "Total Purchase Value": totalrev_age})
purchase_age
```

```python
sn_purchase_data= purchase_data_pd.groupby("SN")
avg_sn= sn_purchase_data["Price"].mean()
purchcount_sn= sn_purchase_data["Purchase ID"].nunique()
totalrev_sn= sn_purchase_data["Price"].sum()
purchase_sn= pd.DataFrame({"Purchase Count": purchcount_sn, "Average Purchase Price": avg_sn, "Total Purchase Value": totalrev_sn})
purchase_sn.head()
```

```python
item_list= purchase_data_pd[["Item ID", "Item Name", "Price", "Purchase ID"]]
item_group= item_list.groupby(["Item ID", "Item Name"])
purchcount_item= item_group["Purchase ID"].nunique()
totalrev_item= item_group["Price"].sum()
single_price= totalrev_item/purchcount_item
purchase_item= pd.DataFrame({"Purchase Count": purchcount_item, "Item Price": single_price, "Total Purchase Value": totalrev_item})
mostpop_item = purchase_item.sort_values(["Purchase Count"], ascending=False)
mostpop_item.head()
```
